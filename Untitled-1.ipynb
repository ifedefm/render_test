```python
import requests
import json

def nuevo_usuario(nueva_contrasenia,nuevo_usuario):
    session = requests.Session()

    url_login = 'https://wallet.casinoenvivo.club/api/admin/login'
    payload_login = {
        'alias': 'Cess93',
        'password': 'Dota2dota2**',
        'otp': ''
    }
    response_login = session.post(url_login, json=payload_login)
    session_id = response_login.json()['user']['session']
    company_id = response_login.json()['user']['company']

    url_register = 'https://wallet.casinoenvivo.club/api/admin/register'
    user_payload = {
        "popupTitle": "",
        "errors": "",
        "currency": "ARS",
        "code_send": False,
        "code_verified": False,
        "alias": nuevo_usuario,
        "password": nueva_contrasenia,
        "national_id": nuevo_usuario,
        "national_id_type": "",
        "email": f"{nuevo_usuario}@mail.com",
        "first_name": nuevo_usuario,
        "last_name": nuevo_usuario,
        "mobile": "1133334444",
        "comments": ""
    }

    payload_register = {
        'session': session_id,
        'user': json.dumps(user_payload),
        'company': company_id
    }

    response_register = session.post(
        url_register,
        data=json.dumps(payload_register),
        headers={"Content-Type": "application/json"}
    )


    res_json = response_register.json()
    print(res_json)
    if res_json.get('result') == 'OK':
        return(f' Usuario creado: {nuevo_usuario}')
    elif "Duplicated alias" in res_json.get('description', ''):
        return(f'Usuario duplicado: {nuevo_usuario}')
    else:
        return(f'Error: {res_json.get("description")}')

            
```

```python
def user_is_valid(usuario_name):
    session = requests.Session()

    url = 'https://wallet.casinoenvivo.club/api/admin/login'
    payload = { 'alias': 'Cess93',
                'password': 'Dota2dota2**',
                'otp': '' }

    response = session.post(url, json=payload)

    response.json()
    session_id = response.json()['user']['session']
    company_id = response.json()['user']['company']
    if session:                             
        url_usuarios = 'https://wallet.casinoenvivo.club/api/admin/agentsUsersSearch'

        payload_users = {
        'session': f'{session_id}',
        'freeText': f'{usuario_name}',
        'company': 'GECN',
        }

        response_user = session.post(url_usuarios, json=payload_users)
        response_user.json()
        dict_user = response_user.json()['users']
        if dict_user == []:
            return False, response_user.json()
        name_user = dict_user[0]['alias']
        if name_user == usuario_name:
            return True, response_user.json()
        else:
            return False, response_user.json()
```
